```python
# Importing numpy and pandas
import numpy as np
import pandas as pd
```

```python
# Assigning datasel url
url='https://raw.githubusercontent.com/tubameric/PortfolioProjects/main/Trivalley%20Customer%20Spend%20and%20Buyer%20Flag%20Data.csv'

# Importing the data into a variable
data = pd.read_csv(url)

# Displaying the data
data
```

```python
#Importing datetime package
import pandas as pd
from datetime import datetime

# Add Age as a new column
data['BirthDate'] = pd.to_datetime(data['BirthDate'])

def calculate_age(born):
    today = datetime.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

# Calculate Age based on the Birth Date
data['Age'] = data['BirthDate'].apply(calculate_age)
```

```python
# Displaying the data columns
data.columns
```

```python
# Setting the dependent variable
y = 'AveMonthSpend'
```

```python
# Setting the noise/redundant variables & variables with more than 20% missing data
ignored_cols = ['CustomerID','Title', 'FirstName','MiddleName', 'LastName', 'Suffix', 'AddressLine1', 'AddressLine2',
              'PhoneNumber', 'BirthDate','TotalChildren']
```

```python
# Setting the categorical variables
cat_cols = ['City', 'StateProvinceName', 'CountryRegionName', 'PostalCode',
       'Education', 'Occupation', 'Gender','MaritalStatus', 'HomeOwnerFlag']
```

```python
# Setting the numerical variables
num_cols = ['NumberChildrenAtHome','YearlyIncome','Age']
```

```python
# Importing the regression libraries from pycaret
from pycaret.regression import *
```

```python
# Setting/configuring the pycaret ML experiment
regression_setup = setup(data,
                        target = y,
                        categorical_features = cat_cols,
                        numeric_features = num_cols,
                        ignore_features = ignored_cols)
```

```python
# Invoking ML algorithms
compare_models()
```

```python
# Building the model using the best Algo
model = create_model('lightgbm')
```

```python
# Using the algo to predict values (avg monthly spend)
predicted_avemonthlyspend = predict_model(model, data)
```

```python
# Downloading the predicted avg monthly spend
predicted_avemonthlyspend.to_csv("Bike Store Predicted Ave Monthly Spend.csv")
```
